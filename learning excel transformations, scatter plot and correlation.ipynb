```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#read the data
df = pd.read_excel(r'C:\Users\My PC\Downloads\scrape\2014\20144.xlsx')

#data test
df.head(5)
```

```python
df.dtypes
```

```python
#convert to string the left 4 to extract year from date
df['year'] = df['dateCreated'].astype(str).str[:4]
df.head(5)
```

```python
#sort by rating count descending
df.sort_values(by=['ratingCount'], inplace=False, ascending= False)
```

```python
#drop the duplicates
df.drop_duplicates()
```

```python
#replace (" [ ]) with none
df['actorname'] = df['actorname'].str.replace("\"","", regex=True).str.replace("[","", regex=True).str.replace("]","", regex=True).str.replace(",",", ", regex=True).str.replace("\n"," ", regex=True)
df['genre'] = df['genre'].str.replace("\"","", regex=True).str.replace("[","", regex=True).str.replace("]","", regex=True).str.replace(",",", ", regex=True).str.replace("\n"," ", regex=True)
df['directorname'] = df['directorname'].str.replace("\"","").str.replace("[","", regex=True).str.replace("]","", regex=True).str.replace(",",", ", regex=True).str.replace("\n"," ", regex=True)
df['creatorname'] = df['creatorname'].str.replace("\"","", regex=True).str.replace("[","", regex=True).str.replace("]","", regex=True).str.replace(",",", ", regex=True).str.replace("\n"," ", regex=True)
df.head(5)
```

```python
#drop actor and image columns
df = df.drop('image', axis=1)
df = df.drop('actor', axis=1)
```

```python
#change ratingCount data type from float to int
df['ratingCount'] = df['ratingCount'].fillna(0).astype('int64')
df.head(5)
```

```python
len(df.index)
```

```python
#remove rows that contains null values
df = df.dropna()
```

```python
len(df.index)
```

```python
df.head(5)
```

```python
#change year data type from object to int
df['year'] = df['year'].fillna(0).astype('int64')
df.dtypes
```

```python
#scatter plot with budget vs gross
plt.scatter(x= df['ratingValue'], y=df['ratingCount'])
plt.title('votes per rate')
plt.xlabel('rate')
plt.ylabel('votes')
plt.show()
```

```python
#plot budget vs gross using seaborn
sns.regplot(x='ratingValue', y='ratingCount', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#correlation 
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('correlation matric for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```
