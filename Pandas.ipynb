```python
#percentage of people from ech country that know python
df_python['PctknowPython'] = (df_python['pythonDevlopers']/df_python['NumberRespondents']) * 100
df_python
```

```python
#sort the results
df_python.sort_values(by='PctknowPython', ascending=False, inplace=True)
df_python
```
