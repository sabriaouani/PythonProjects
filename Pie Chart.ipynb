```python
from matplotlib import pyplot as plt
```

```python
#creating a simple pie chart
plt.tight_layout()
plt.style.use('seaborn')
slices = [120, 80, 30, 20]
labels = ['sabri', 'youssef','ahmed','achref']
plt.title('My awesome pie chart')
plt.pie(slices, labels=labels)
```

```python
#you can add wadge with wadgeprops, search matplotlib wadges for more custom features, i will change also the colors
colors = ['blue','red','yellow','green']
plt.pie(slices, labels=labels, wedgeprops= {'edgecolor': 'red'}, colors = colors)
```

```python
# Language Popularity
nbrResponses = [59219, 55466, 47544, 36443, 35917]
languages = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(nbrResponses, labels=languages, wedgeprops= {'edgecolor': 'red'})
```

```python
#explode one of the pie out of the pie chart usung 'explode'
explode = [0,0,0,0.1,0]
plt.pie(nbrResponses, labels=languages, wedgeprops= {'edgecolor': 'red'}, explode=explode)
```

```python
#adding a shodow to the pie chart
plt.pie(nbrResponses, labels=languages, wedgeprops= {'edgecolor': 'black'}, explode=explode, shadow= True)
```

```python
# to rotate the chart use 'startangle' and a percentage (javascript started at 90 degree)
plt.pie(nbrResponses, labels=languages, wedgeprops= {'edgecolor': 'black'}, explode=explode, shadow= True, startangle=90)
```

```python
#to display the percentage each slice has use
plt.pie(nbrResponses, labels=languages, wedgeprops= {'edgecolor': 'black'}, 
        explode=explode, shadow= True, startangle=90, autopct='%1.1f%%')
```
