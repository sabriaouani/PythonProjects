```python
#adding a filling area
plt.plot(ages_x, js_dev_y ,  linewidth=1 , label='JavaScript')
plt.plot(ages_x, dev_y , label='All Devs')
plt.plot(ages_x, py_dev_y, label='Python')
plt.legend()
#method. P.S to see through, use alpha
plt.fill_between(ages_x, js_dev_y,alpha= 0.25)
```

```python
#instead of ploting a 0, you can use a median, now is showing us where exactly our python plot crossed that median
median= 74000
plt.plot(ages_x, js_dev_y ,  linewidth=1 , label='JavaScript')
plt.plot(ages_x, dev_y , label='All Devs')
plt.plot(ages_x, py_dev_y, label='Python')
plt.legend()
plt.fill_between(ages_x, py_dev_y, median ,alpha= 0.25)
```

```python
#interpolate will make sure that certain sections dont get cliped and all the regions are filled correctly
# using where conding will show uper half with a color and the second with a color
median= 74000
plt.plot(ages_x, js_dev_y , label='JavaScript')
plt.plot(ages_x, dev_y , label='All Devs')
plt.plot(ages_x, py_dev_y, label='Python')
plt.legend()
plt.fill_between(ages_x, dev_y, median , where = (dev_y > median) ,  interpolate=True, alpha= 0.25, label='above avg')
plt.fill_between(ages_x, dev_y, median , where = dev_y <= median ,  interpolate=True, alpha= 0.25, label='below avg')
```
