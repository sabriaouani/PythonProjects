```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
#building a simple scatter plot, 's=100' will make the dots larger 
plt.scatter(x, y, s=100)
```

```python
#in this link you will find marker styles: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.scatter(x, y, s=100, color='green', marker='x')
```

```python
#to make look better use
plt.scatter(x, y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.75)
```

```python
#to assign different colors to the plot, so x5 y5 is having color 7, x7 y4 having color 4 ans so on
plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75)
```

```python
#you can find color map (shades) colors under this link: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
plt.scatter(x, y, s=100, c=colors, cmap='Greens' , edgecolor='black', linewidth=1, alpha=0.75)
```

```python
#to get color legend bar use:
plt.scatter(x, y, s=100, c=colors, cmap='Greens' , edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

```python
#using sizes list
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens' , edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

```python
#working with real world data
data = pd.read_csv('views.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
#building a scratter plot with real world data
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)
plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total likes')
```

```python
#for better view, im gonna use log scale for x and y axis instead of regular scale
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')
```

```python
#to use ratio for colors
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Ratio')
plt.xscale('log')
plt.yscale('log')
plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total likes')
```
